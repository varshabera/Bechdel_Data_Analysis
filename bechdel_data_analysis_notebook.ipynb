# Bechdel Test Movie Data Analysis

This notebook performs exploratory data analysis on the Bechdel Test movie dataset.

Steps included:
- Load dataset
- Data overview
- Data cleaning
- Exploratory data analysis
- Visualization
- Correlation analysis
- Insights

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## Load Dataset

```python
df = pd.read_csv('cleaned_bechdel_data.csv')
df.head()
```

## Dataset Overview

```python
print('Dataset Shape:', df.shape)
print('\nColumns:')
print(df.columns)

df.info()
```

## Summary Statistics

```python
df.describe()
```

## Check Missing Values

```python
df.isnull().sum()
```

## Remove Duplicates

```python
df = df.drop_duplicates()
```

## Bechdel Score Distribution

```python
df['bechdel_rating'].value_counts()
```

## Bechdel Score Visualization

```python
df['bechdel_rating'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of Bechdel Scores')
plt.xlabel('Bechdel Score')
plt.ylabel('Number of Movies')
plt.show()
```

## Create Pass/Fail Column

```python
df['Pass_Test'] = df['bechdel_rating'].apply(lambda x: 'Pass' if x == 3 else 'Fail')
df.head()
```

## Pass vs Fail Count

```python
df['Pass_Test'].value_counts()
```

## Pass vs Fail Visualization

```python
sns.countplot(data=df, x='Pass_Test')
plt.title('Bechdel Test Results')
plt.show()
```

## IMDb Rating Analysis

```python
df.groupby('Pass_Test')['imdb_rating'].mean()
```

## IMDb Rating Visualization

```python
sns.boxplot(x='Pass_Test', y='imdb_rating', data=df)
plt.title('IMDb Rating vs Bechdel Result')
plt.show()
```

## Runtime Analysis

```python
df.groupby('Pass_Test')['runtime'].mean()
```

## Runtime Visualization

```python
sns.boxplot(x='Pass_Test', y='runtime', data=df)
plt.title('Runtime vs Bechdel Result')
plt.show()
```

## Correlation Analysis

```python
numeric_cols = ['bechdel_rating','imdb_rating','runtime']
corr = df[numeric_cols].corr()
corr
```

## Correlation Heatmap

```python
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

## Insights

- Majority of movies pass the Bechdel Test.
- IMDb ratings between pass and fail movies are very similar.
- Runtime does not significantly influence Bechdel results.
- Correlation between Bechdel score and other variables is weak.
